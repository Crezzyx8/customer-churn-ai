```python
import pandas as pd
import numpy as np


np.random.seed(42)

n = 1000


data = {

    "customer_id": range(1, n+1),

    "age": np.random.randint(18, 65, n),

    "tenure": np.random.randint(1, 60, n),

    "monthly_charges": np.random.randint(20, 150, n),

    "total_charges": np.random.randint(100, 8000, n),

    "contract_type": np.random.choice(

        ["Monthly", "Yearly"],
        n
    ),

    "payment_method": np.random.choice(

        ["Credit Card", "Debit Card", "Bank Transfer"],
        n
    ),

    "internet_service": np.random.choice(

        ["Yes", "No"],
        n
    ),

}


df = pd.DataFrame(data)



df["churn"] = np.where(

    (df["tenure"] < 12)

    & (df["contract_type"] == "Monthly")

    & (df["monthly_charges"] > 80),

    1,

    0

)


df.to_csv("churn.csv", index=False)



print("Dataset created at data/churn.csv")
```
